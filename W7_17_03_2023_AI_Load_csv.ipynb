```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import linear_model
url = "mtcars.csv"
data = pd.read_csv(url)
data.head()
```

```python
data.hist()
plt.show()
sns.pairplot(data)
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='red')
```

```python
from sklearn import linear_model
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']), y=data['mpg'])    
```

```python
print('Gi tr intercept:', model.intercept_)
print('Gi tr coeffient:', model.coef_)
print(' chnh xc:', model.score(pd.DataFrame(data['wt']),y=data['mpg']))
```

```python
data.plot(kind = 'scatter',x='wt',y='mpg',figsize=(9,9),color = 'black',xlim = (0,7))
y_pred = model.predict(X=pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color = 'orange')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame({'wt': data['wt'], 'wt_sq': data['wt']**2})
model.fit(X,y=data['mpg'])
print('Gi tr intercept:', model.intercept_)
print('Gi tr coeffient:', model.coef_)
print(' chnh xc:', model.score(X,y=data['mpg']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import linear_model
url = "50_Startups.csv"
data = pd.read_csv(url)
data.head()
```

```python
data.plot(kind ='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color = 'black',xlim = (0,200000))
y_pred = model.predict(X=pd.DataFrame(data['Profit']))
plt.plot(data['Profit'],y_pred,color ='red')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame({'Profit': data['Profit'], 'Profit_sq': data['Profit']**2})
model.fit(X,y=data['R&D Spend'])
print('Gi tr intercept:', model.intercept_)
print('Gi tr coeffient:', model.coef_)
print(' chnh xc:', model.score(X,y=data['R&D Spend']))
```

```python
range = np.arange(1.5,5.5,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter',x='Profit',y='R&D Spend',figsize=(9,9),color = 'black',xlim = (0,300000))
plt.plot(range,y_pred,color = 'blue')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import linear_model
url = "Advertising.csv"
data = pd.read_csv(url)
data.head()
correlation = data.corr()
print(correlation)
sns.heatmap(correlation, annot = True)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['sales']), y=data['TV']) 
print('Gi tr intercept:', model.intercept_)
print('Gi tr coeffient:', model.coef_)
print(' chnh xc:', model.score(pd.DataFrame(data['sales']),y=data['TV']))
```

```python
data.plot(kind ='scatter',x='sales',y='TV',figsize=(9,9),color = 'black',xlim = (0,100))
y_pred = model.predict(X=pd.DataFrame(data['sales']))
plt.plot(data['sales'],y_pred,color ='red')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame({'sales': data['sales'], 'sales_sq': data['sales']**2})
model.fit(X,y=data['TV'])
print('Gi tr intercept:', model.intercept_)
print('Gi tr coeffient:', model.coef_)
print(' chnh xc:', model.score(X,y=data['TV']))
range = np.arange(1,30,0.01)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter',x='sales',y='TV',figsize=(9,9),color = 'black',xlim = (0,200))
plt.plot(range,y_pred,color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = data.loc[:, ['TV', 'newspaper']]
y = data['sales']
model.fit(X, y)

print('Gi tr intercept:', model.intercept_)
print('Gi tr coeffient:', model.coef_)
print(' chnh xc:', model.score(X, y))
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['TV'], data['sales'], label='TV')
plt.scatter(data['newspaper'], data['sales'], label='newspaper')
plt.xlabel('Qung co trn TV v bo')
plt.ylabel('Doanh s bn hng')
plt.legend()
plt.show()
```
